```python
ref_gse_df = df_shci_1[['system', 'basis', 'SHCI']]
ref_gse_df
```

```python
shci_ref_db = pd.concat([ref_gse_df, ref_gse_df_3], ignore_index=True)

# Function to check if the entry is a float, convert to string if true, then concatenate
def process_row(row):
    if not pd.isna(row['charge']):  # Check if the entry is a float
        return f"{row['system']}_{str(row['charge'])}"  # Convert to string and concatenate
    else:
        return row['system']  # Return the first column entry as is

# Apply the function to each row
shci_ref_db['system'] = shci_ref_db.apply(process_row, axis=1)

shci_ref_db = shci_ref_db.drop(columns='charge')

shci_ref_db
```
